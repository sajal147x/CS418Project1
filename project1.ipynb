```python
import pandas as pd
```

```python
try:
    from google.colab import files

    print("Upload demographics file:")
    uploaded = files.upload()
    demographics = pd.read_csv(list(uploaded.keys())[0])

    print("Upload election file:")
    uploaded = files.upload()
    election = pd.read_csv(list(uploaded.keys())[0])

except ImportError:
    demographics = pd.read_csv("demographics_data.csv")
    election = pd.read_csv("election_data.csv")
```

```python
election = election.pivot_table(
    index=["YEAR", "STATE", "COUNTY"],
    columns="PARTY",
    values="VOTES",
    aggfunc="sum"
).reset_index()
```

```python
print(f" THE ROW SIZE: {election.shape[0]}")
print(f" NUM COLUMNS: {election.shape[1]}")
print()
```

```python
state_to_abbrev = {
    'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR',
    'California': 'CA', 'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE',
    'Florida': 'FL', 'Georgia': 'GA', 'Hawaii': 'HI', 'Idaho': 'ID',
    'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA', 'Kansas': 'KS',
    'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD',
    'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS',
    'Missouri': 'MO', 'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV',
    'New Hampshire': 'NH', 'New Jersey': 'NJ', 'New Mexico': 'NM', 'New York': 'NY',
    'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH', 'Oklahoma': 'OK',
    'Oregon': 'OR', 'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC',
    'South Dakota': 'SD', 'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT',
    'Vermont': 'VT', 'Virginia': 'VA', 'Washington': 'WA', 'West Virginia': 'WV',
    'Wisconsin': 'WI', 'Wyoming': 'WY', 'District of Columbia': 'DC'
}
```

```python
demographics['State Name'] = demographics['State Name'].map(state_to_abbrev).fillna(demographics['State Name'])
#making sure it can execute multiple times
```

```python
# manually removing parish since it is there in one dataset not the other, and adding County at the end for all
def fix_county(x):
    if 'Parish' in x:
         x = x.replace(' Parish', '').replace('Parish', '')
    if 'County' in x:
        return x
    else:
        return x + ' County'

demographics['County Name'] = demographics['County Name'].apply(fix_county)
election['COUNTY'] = election['COUNTY'].apply(fix_county)
```

```python
merged = pd.merge(demographics, election, left_on=['State Name', 'County Name'], right_on=['STATE', 'COUNTY'], how='outer')
```
